## 이미지 분석 딥러닝 배경

**기존 MLP의 문제점**
- 한칸씩 픽셀을 움직여도 변화하는 인풋값이 바뀌므로 결과가 확 바뀜
- 때문에 픽셀이 아닌 특징을 찾아서 학습해야 함

**CNN**
- Feature Extraction: 필터가 지나가면서 가운데 정렬을 시키기도 하고 특징을 찾기 때문에 변화이동에 대해 둔감함
- 때문에 이미지 분석에 탁월함

**Multi Object Detection**
- 여러가지 물체에 대한 classification + 물체의 위치 정보를 파악하는 Localization

**1 Stage vs 2 Stage Dtector**

**2-Stage Detector**

- Regional Proposal 과 Classification이 순차적으로 이루어짐
- 즉, 여러 객체 중 해당 객체가 어디있는지 확인하고, 이 객체가 무엇인지 분류함
- 단계가 나눠져 있으므로 시간이 오래걸림

**1-Stage Detector**
- 딥러닝이 한번만 이루어짐
- 특징을 그리드로 나누어서 각각 detection을 시킨뒤 연결시킨다
- **YOLO**: you only look once -> "한번만 보고 다 맞춘다"
